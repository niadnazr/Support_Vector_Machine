```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
col = ['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
for i in col:
     salary_train[i]=le.fit_transform(salary_train[i])
     salary_test[i]=le.fit_transform(salary_test[i]) 
```

```python
from sklearn.preprocessing import LabelEncoder,LabelBinarizer
bnry = LabelBinarizer()
salary_train['Salary'] = bnry.fit_transform(salary_train['Salary'])
salary_test['Salary'] = bnry.fit_transform(salary_test['Salary'])
```

```python
x_train = salary_train.iloc[:,0:13]
y_train = salary_train.iloc[:,13]
x_test = salary_test.iloc[:,0:13]
y_test = salary_test.iloc[:,13]
```

```python
from sklearn import svm
from sklearn.svm import SVC
model = SVC(kernel = 'poly')
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
pred_test_poly = model.predict(x_test)
acc = accuracy_score(y_test,pred_test_poly)
print('ACCURACY',acc)
```

```python
print(confusion_matrix(y_test,pred_test_poly))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_test_poly))
```
